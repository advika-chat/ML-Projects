Every neuron has one single bias (we can turn multiple biases into one by simplifying). 
The number of inputs determine the number of weights. 

Inputs can either be values from the input layer (which is a vector of values) or the outputs from the neurons of a previous hidden layer. 

In the output layer, each neuron will have a separate set of weights and a separate value for the bias, hence a unique output.

To make a model which fits your use better, you can alter the weights and baises to get a better accuracy. Your input cannot be changed since it is raw data that you have obtained and the basis of your neural network. 

```python
inputs = [1 ,2 , 3, 2.5]

weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]

biases = [2, 3, 0.5]

layer_outputs = []
for neuron_weights, neuron_bias in zip(weights, biases):
    neuron_output = 0
    for n_input, weight in zip(inputs, neuron_weights):
        neuron_output += n_input * weight
    neuron_output += neuron_bias
    layer_outputs.append(neuron_output)

print(layer_outputs)
```

Zip function combines two lists into a list of lists, element wise. 
Weights and biases that the optimizer tunes to attempt the results to fit the given data.
