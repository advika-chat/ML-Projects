```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#commonly x and y are 1d arrays
x = [0,1,2,3]
y = [0,2,4,6]

plt.plot(x,y, label = '2x', color = '000000', linewidth = 2, marker='.')

plt.title('First Graph')
plt.xlabel('x axis')

plt.xticks([0, 1, 2, 3])
plt.yticks([0,2,4,6])

plt.legend()

plt.show()
```
